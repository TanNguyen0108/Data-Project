```python
soup_list = []
for i in url_tolist:
    print('https://'+i)
    try:
        soup = BeautifulSoup(requests.get('https://'+i).content, 'lxml')
        soup_list.append(soup)
    except: print(f'Error when accessing url {i}')
```

```python
latitude, longitude, restaurant_id, district_id, city_id, name, street_address, area_address, district_address, region_address, category, cuisine, audience, average_rating, position_rating, price_rating, \
quality_rating, service_rating, space_rating, review_count, review_excellent_count, review_good_count, review_average_count, \
review_bad_count, view_count, checked_in_count, like_count, favourite_count, wanted_count, picture_count, status, time_clock, price_range, url \
= [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], [], []
for i in soup_list:
    if i.find('meta',{'property':'place:location:latitude'}) == None:
        latitude.append(None)
    else: latitude.append(i.find('meta',{'property':'place:location:latitude'}).get('content'))

    if i.find('meta',{'property':'place:location:longitude'}) == None:
        longitude.append(None)
    else: longitude.append(i.find('meta',{'property':'place:location:longitude'}).get('content'))
        
    if i.find('div',{'class':'micro-left-content'}) == None:
        restaurant_id.append(None)
        district_id.append(None)
        city_id.append(None)
    else:
        restaurant_id.append(re.findall('\d+', re.findall('"RestaurantID":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])
        district_id.append(re.findall('\d+', re.findall('"DistrictID":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])
        city_id.append(re.findall('\d+', re.findall('"CityID":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])

    if i.find('h1',{'itemprop':'name'}) == None:
        name.append(None)
    else: name.append(i.find('h1',{'itemprop':'name'}).string)

    if i.find('span',{'itemprop':'streetAddress'}) == None:
        street_address.append(None)
    else: street_address.append(i.find('span',{'itemprop':'streetAddress'}).string)
        
    if i.find('meta',{'content':'3'}) == None:
        area_address.append(None)       
    else: area_address.append(i.find('meta',{'content':'3'}).find_next().find_next().string)

    if i.find('span',{'itemprop':'addressLocality'}) == None:
        district_address.append(None)
    else: district_address.append(i.find('span',{'itemprop':'addressLocality'}).string)

    if i.find('span',{'itemprop':'addressRegion'}) == None:
        region_address.append(None)
    else: region_address.append(i.find('span',{'itemprop':'addressRegion'}).string)

    if i.find('div',{'class':'category-items'}) == None:
        category.append(None)
    else: category.append(i.find('div',{'class':'category-items'}).find_next().string.replace(',',''))

    if i.find('div',{'itemprop':'servesCuisine'}) == None:
        cuisine.append(None)
    else:
        sum = ''
        for j in i.find('div',{'itemprop':'servesCuisine'}).findChildren():
            sum += j.string.replace('\r','').replace('\n','').strip().replace(',',', ')
        cuisine.append(sum)

    if i.find('div',{'class':'audiences'}) == None:
        audience.append(None)
    else: audience.append(i.find('div',{'class':'audiences'}).string.replace('\r',' ').replace('\n',' ').replace(' ',' ').replace('\xa0',' ').replace('-','').strip())
    
    if i.find('div',{'class':'microsite-point-avg'}) == None or i.find('div',{'class':'microsite-point-avg'}).string == None:
        average_rating.append(None)
    else: average_rating.append(i.find('div',{'class':'microsite-point-avg'}).string.replace('\r','').replace('\n','').replace(' ',''))

    if len(i.findAll('div',{'class':'microsite-top-points'})) == 0:
        position_rating.append(None)
        price_rating.append(None)
        quality_rating.append(None)
        service_rating.append(None)
        space_rating.append(None)
    else:
        point, label = [], []
        for j in i.findAll('div',{'class':'microsite-top-points'}):
            point.append(j.find_next().find_next().string)
            label.append(j.find_next().find_next().find_next().string)

        label_point = {label[j]: point[j] for j in range(len(label))}

        position_rating.append(label_point['Vị trí'])
        price_rating.append(label_point['Giá cả'])
        quality_rating.append(label_point['Chất lượng'])
        service_rating.append(label_point['Phục vụ'])
        space_rating.append(label_point['Không gian'])

    if i.find('div',{'class':'micro-left-content'}) == None:
        review_count.append(None)
    else: review_count.append(re.findall('\d+', re.findall('"TotalReview":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])

    if len(i.findAll('span',{'class':'rating-levels'})) == 0:
        review_excellent_count.append(None)
        review_good_count.append(None)
        review_average_count.append(None)
        review_bad_count.append(None)
    else:
        review_excellent_count.append(i.findAll('span',{'class':'rating-levels'})[0].find_next().string.replace('\r','').replace('\n','').strip())
        review_good_count.append(i.findAll('span',{'class':'rating-levels'})[1].find_next().string.replace('\r','').replace('\n','').strip())
        review_average_count.append(i.findAll('span',{'class':'rating-levels'})[2].find_next().string.replace('\r','').replace('\n','').strip())
        review_bad_count.append(i.findAll('span',{'class':'rating-levels'})[3].find_next().string.replace('\r','').replace('\n','').strip())

    if i.find('div',{'class':'micro-left-content'}) == None:
        view_count.append(None)
    else: view_count.append(re.findall('\d+', re.findall('"TotalView":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])

    if i.find('div',{'class':'micro-left-content'}) == None:
        like_count.append(None)
    else: like_count.append(re.findall('\d+', re.findall('"TotalLikes":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])

    if i.find('div',{'class':'micro-left-content'}) == None:
        checked_in_count.append(None)
    else: checked_in_count.append(re.findall('\d+', re.findall('"TotalCheckedIn":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])

    if i.find('div',{'class':'micro-left-content'}) == None:
        favourite_count.append(None)
    else: favourite_count.append(re.findall('\d+', re.findall('"TotalFavourite":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])

    if i.find('div',{'class':'micro-left-content'}) == None:
        wanted_count.append(None)
    else: wanted_count.append(re.findall('\d+', re.findall('"TotalWanted":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])

    if i.find('div',{'class':'micro-left-content'}) == None:
        picture_count.append(None)
    else: picture_count.append(re.findall('\d+', re.findall('"TotalPictures":\d+,', i.find('div',{'class':'micro-left-content'}).find_next().find_next().find_next().string)[0])[0])

    if i.find('span',{'class':'fa fa-clock-o houricon'}) == None:
        status.append(None)
    else: status.append(i.find('span',{'class':'fa fa-clock-o houricon'}).find_next().string)

    if i.find('span',{'class':'fa fa-clock-o houricon'}) == None:
        time_clock.append(None)
    elif i.find('span',{'class':'fa fa-clock-o houricon'}).find_next().find_next().get('ng-click') == None:
        time_clock.append(i.find('span',{'class':'fa fa-clock-o houricon'}).find_next().find_next().string.replace(' ','').replace('\xa0',''))
    else: time_clock.append(None)

    if i.find('span',{'class':'fa fa-tag minmaxpriceicon'}) == None:
        price_range.append(None)
    elif i.find('span',{'itemprop':'priceRange'}) == None:
        price_range.append(i.find('span',{'class':'fa fa-tag minmaxpriceicon'}).find_next().string)
    else: price_range.append(i.find('span',{'itemprop':'priceRange'}).get_text().replace('\n',''))

    if i.find('meta',{'property':'og:url'}) == None:
        url.append(None)
    else: url.append(i.find('meta',{'property':'og:url'}).get('content'))   
```
